## 2. Les principales méthodes utilisées dans un CNN

Dans le programme étudié, on utilise plusieurs couches.  
Voici les plus importantes.



### 2.1. `Input()`

```python
layers.Input(shape=(28,28,1))
```

Cette instruction définit la forme des données d’entrée.

Ici :

- `28` : hauteur de l’image ;
- `28` : largeur de l’image ;
- `1` : nombre de canaux.

Comme les images de MNIST sont en niveaux de gris, il n’y a qu’un seul canal.

---

### 2.2. `Conv2D()`

```python
layers.Conv2D(32, (3,3), activation="relu")
```

Cette couche applique des **filtres** sur l’image.

Rôle :
- détecter des motifs simples ;
- par exemple des bords, des segments, des courbes.

Dans l’exemple :

- `32` = nombre de filtres ;
- `(3,3)` = taille de chaque filtre ;
- `activation="relu"` = fonction d’activation.

---

### 2.3. `MaxPooling2D()`

```python
layers.MaxPooling2D((2,2))
```

Cette couche réduit la taille des données.

Rôle :
- diminuer le nombre de calculs ;
- conserver l’information essentielle ;
- rendre le modèle plus robuste à de petits décalages.

Le principe est simple : dans chaque bloc `2 × 2`, on garde la valeur maximale.

---

### 2.4. `Flatten()`

```python
layers.Flatten()
```

Cette couche transforme les données en vecteur.

Avant `Flatten()`, on a encore une structure de type image ou tableau multidimensionnel.  
Après `Flatten()`, toutes les valeurs sont placées dans une seule liste.

Cette transformation est nécessaire avant une couche `Dense()`.

---

### 2.5. `Dense()`

```python
layers.Dense(64, activation="relu")
```

Une couche `Dense` est une couche **entièrement connectée**.

Cela signifie que chaque neurone de cette couche reçoit toutes les sorties de la couche précédente.

Rôle :
- combiner les caractéristiques extraites par les convolutions ;
- préparer la décision finale.

---

### 2.6. `Dropout()`

```python
layers.Dropout(0.3)
```

Cette couche désactive aléatoirement une partie des neurones pendant l’entraînement.

Ici :

- `0.3` signifie que 30 % des neurones sont temporairement éteints.

Rôle :
- éviter qu’un petit groupe de neurones devienne trop important ;
- limiter le surapprentissage ;
- améliorer la généralisation.



## 3. Critères de sélection d’un modèle

Pour comparer plusieurs modèles, on observe quatre indicateurs principaux :

- `accuracy`
- `loss`
- `val_accuracy`
- `val_loss`



### 3.1. `accuracy`

L’accuracy mesure la proportion de bonnes réponses.

accuracy = bonnes prédictions / nombre total de prédictions

C’est donc un **pourcentage de réussite**.

---

### 3.2. `loss`

La `loss` mesure l’erreur du modèle.

Attention : ce n’est **pas** un simple pourcentage d’erreur.  
Elle tient compte de la **confiance** du modèle dans sa réponse.

Un modèle peut avoir raison, mais avec une faible probabilité : dans ce cas, la `loss` reste assez grande.

*Exemple : si la valeur "2" sort mais que P(2) = 0.3 => loss élevée (alors que si P(2) = 0.8 => loss faible).*

---

### 3.3. `val_accuracy`

La `val_accuracy` est l’accuracy calculée sur les données de validation.

Ces données ne servent pas directement à l’apprentissage.

Elles permettent donc de savoir si le modèle reste bon sur des données qu’il n’a pas apprises par cœur.

---

### 3.4. `val_loss`

La `val_loss` est la loss calculée sur les données de validation.

C’est souvent le critère le plus important pour choisir un modèle.

Pourquoi ?

Parce qu’elle permet de repérer le **surapprentissage** :

- si la `loss` d’entraînement continue de baisser ;
- mais que la `val_loss` remonte ;

alors le modèle devient trop spécialisé sur les données d’entraînement.



## 4. Méthode expérimentale
